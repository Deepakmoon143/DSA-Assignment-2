Question_1 ; Delete the elements in an linked list whose sum is equal to zero

```python
class List:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_lst = List(data)
        if self.head is None:
            self.head = new_lst
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_lst

    def delete_zero_sum (self):
        current = self.head
        while current:
            sum = 0
            temp = current
            while temp:
                sum += temp.data
                temp = temp.next
                if sum == 0:
                    break
            if sum == 0:
                self.head = current.next
            else:
                temp = current
                prev = None
                while temp:
                    sum += temp.data
                    if sum == 0:
                        prev.next = temp.next
                        break
                    prev = temp
                    temp = temp.next
            current = current.next

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" --> ")
            current = current.next
        print("None")


linked_list = LinkedList()
n = int(input("Enter the number of elements: "))
for i in range(n):
    data = int(input(f"Enter element {i + 1}: "))
    linked_list.insert(data)

print("Original Linked List:")
linked_list.display()

linked_list.delete_zero_sum()

print("Linked List after deleting zero sum nodes:")
linked_list.display()
```

Question_2 ; Reverse a linked list in groups of given size

```python
class List:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_lst = List(data)
        if self.head is None:
            self.head = new_lst 
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_lst 

    def reverse(self, head, k):
        if not head:
            return None

        current = head
        next_lst  = None
        prev = None
        count = 0

        
        while current and count < k:
            next_lst  = current.next
            current.next = prev
            prev = current
            current = next_lst 
            count += 1

        if next_lst :
            head.next = self.reverse(next_lst , k)

        return prev

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" --> ")
            current = current.next
        print("None")

linked_list = LinkedList()
n = int(input("Enter the number of elements: "))
for i in range(n):
    data = int(input(f"Enter element {i + 1}: "))
    linked_list.insert(data)

k = int(input("Enter the group size for reversing: "))

print("Original Linked List:")
linked_list.display()


linked_list.head = linked_list.reverse(linked_list.head, k)

print(f"Linked List after reversing in groups of {k}:")
linked_list.display()
```

Question_3 ; Merge a linked list into another linked list at alternate positions.

```python
class List:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_lst = List(data)
        if self.head is None:
            self.head = new_lst
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_lst

    def merge(self, other_list):
        current1 = self.head
        current2 = other_list.head

        while current1 and current2:
            next1 = current1.next
            next2 = current2.next

            current1.next = current2
            current2.next = next1

            current1 = next1
            current2 = next2

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


linked_list1 = LinkedList()
n1 = int(input("Enter the number of elements for the first linked list: "))
for i in range(n1):
    data = int(input(f"Enter element {i + 1} for the first linked list: "))
    linked_list1.insert(data)

linked_list2 = LinkedList()
n2 = int(input("Enter the number of elements for the second linked list: "))
for i in range(n2):
    data = int(input(f"Enter element {i + 1} for the second linked list: "))
    linked_list2.insert(data)

print("First Linked List:")
linked_list1.display()
print("Second Linked List:")
linked_list2.display()

linked_list1.merge(linked_list2)

print("Merged Linked List:")
linked_list1.display()
```

Question_4 ; In an array, Count Pairs with given sum.

```python
def count_pairs(arr, target_sum):
    count = 0
    n = len(arr)

    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == target_sum:
                count += 1

    return count


def input_array():
    elements = input("Enter elements of the array separated by spaces: ").split()
    return list(map(int, elements))


arr = input_array()
target_sum = int(input("Enter the target sum: "))

result = count_pairs(arr, target_sum)
print(f"Number of pairs with a sum of {target_sum}: {result}")
```

Question_5 ; Find duplicates in an array.

```python
def find_duplicates(arr):
    duplicates = []
    seen = set()

    for num in arr:
        if num in seen:
            duplicates.append(num)
        else:
            seen.add(num)

    return duplicates

def input_array():
    elements = input("Enter elements of the array separated by spaces: ").split()
    return list(map(int, elements))


arr = input_array()

duplicate_elements = find_duplicates(arr)

if duplicate_elements:
    print("Duplicates in the array:", duplicate_elements)
else:
     print("No duplicates found in the array.")
```

Question_6 ; Find the Kth largest and Kth smallest number in an array.

```python
def kth_largest_smallest(arr, k):
    sorted_arr = sorted(arr)
    
    if k > 0 and k <= len(sorted_arr):

        kth_largest = sorted_arr[-k]
        kth_smallest = sorted_arr[k - 1]

        return kth_largest, kth_smallest
    
    else:
        return None, None


def input_array():
    elements = input("Enter elements of the array separated by spaces: ").split()
    return list(map(int, elements))


arr = input_array()
k = int(input("Enter the value of K: "))

kth_largest, kth_smallest = kth_largest_smallest(arr, k)

if kth_largest is not None and kth_smallest is not None:
    print(f"{k}th largest number: {kth_largest}")
    print(f"{k}th smallest number: {kth_smallest}")
else:
    print("Invalid value of K. Please enter a valid value.")
```

Question_7 ; Move all the negative elements to one side of the array.

```python
def negatives(arr):
    n = len(arr)
    left = 0
    right = n - 1

    while left <= right:
        if arr[left] < 0 and arr[right] < 0:
            left += 1
        elif arr[left] >= 0 and arr[right] < 0:
            left += 1
            right -= 1
        elif arr[left] < 0 and arr[right] >= 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        else:
            right -= 1


def input_array():
    elements = input("Enter elements of the array separated by spaces: ").split()
    return list(map(int, elements))


arr = input_array()

print("Original Array:", arr)
negatives(arr)
print("Array with Negative Elements on One Side:", arr)
```

Question_8 ; Reverse a string using a stack data structure

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

def reverse_string(input_str):
    stack = Stack()
    for i in input_str:
        stack.push(i)

    reversed_str = ""
    while not stack.is_empty():
        reversed_str += stack.pop()

    return reversed_str

input_str = input("Enter a string: ")

reversed_str = reverse_string(input_str)

print("Original String:", input_str)
print("Reversed String:", reversed_str)
```

Question_9 ; Evaluate a postfix expression using stack

```python
def evaluate_postfix(expression):
    stack = []
    operators = set(['+', '-', '*', '/'])

    for i in expression:
        if i.isdigit():
            stack.append(int(i))
        elif i in operators:
            if len(stack) < 2:
                return "Invalid expression"

            operand2 = stack.pop()
            operand1 = stack.pop()

            if i == '+':
                result = operand1 + operand2
            elif i == '-':
                result = operand1 - operand2
            elif i == '*':
                result = operand1 * operand2
            elif i == '/':
                if operand2 == 0:
                    return "Division by zero error"
                result = operand1 / operand2

            stack.append(result)

    if len(stack) == 1:
        return stack[0]
    else:
        return "Invalid expression"


postfix_expression = input("Enter a postfix expression: ")

result = evaluate_postfix(postfix_expression)

print("Result of the postfix expression:", result)

```

Question_10 ; Implement a queue using the stack data structure

```python
class QueueUsingList:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"
        return self.queue.pop(0)


def operations():
    queue = QueueUsingList()
    while True:
        print("\nQueue Operations:")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            item = input("Enter the item to enqueue: ")
            queue.enqueue(item)
            print(f"{item} enqueued to the queue.")
        elif choice == '2':
            item = queue.dequeue()
            if item:
                print(f"{item} dequeued from the queue.")
            else:
                print("Queue is empty.")
        elif choice == '3':
            break
        else:
            print("Invalid choice. Please try again.")


print("Queue using a List")
operations()
```
